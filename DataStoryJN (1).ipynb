```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

# Assuming 'new_df' is your DataFrame containing the data
# Assuming 'Class' is the target variable

# Splitting data into features (X) and target variable (y)
X = new_df[['Atr18', 'Atr20', 'Atr39']]
y = new_df['Class']

# Splitting data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for training data
ax.scatter(X_train['Atr18'], X_train['Atr20'], X_train['Atr39'], c=y_train, marker='o', label='Training Data')

# Scatter plot for test data
ax.scatter(X_test['Atr18'], X_test['Atr20'], X_test['Atr39'], c='g', marker='o', label='Test Data')

# Set labels and title
ax.set_xlabel('Answer for 18')
ax.set_ylabel('Answer for 20')
ax.set_zlabel('Answer for 39')
ax.set_title('Clustering Plot')

# Perform K-means clustering on training data
kmeans = KMeans(n_clusters=2)
kmeans.fit(X_train)
centroids = kmeans.cluster_centers_

# Plot centroids
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], c='r', marker='X', s=200, label='Centroids')

# Show legend
ax.legend()

# Show plot
plt.show()
```

```python
from sklearn.metrics import pairwise_distances_argmin_min

# Predicting classes for test data based on closest centroid
closest_centroids, _ = pairwise_distances_argmin_min(X_test, centroids)
predicted_class = closest_centroids

# Adding predicted class to test set
X_test_with_pred = X_test.copy()
X_test_with_pred['predicted_class'] = predicted_class

# Calculating accuracy
accuracy = (X_test_with_pred['predicted_class'] == y_test).mean()
print("Accuracy:", accuracy)
```
